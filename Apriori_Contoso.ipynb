```python
# Adicionando o apriori da biblioteca
from apyori import apriori
import pandas as pd
```

```python
# Exemplo de transações (lista de listas, onde cada lista representa uma transação)
transacoes = [
    ['leite', 'pão', 'manteiga'],
    ['leite', 'fraldas', 'cerveja', 'arroz'],
    ['pão', 'fraldas', 'cerveja', 'pizza'],
    ['pão', 'fraldas', 'cerveja', 'leite'],
    ['pão', 'leite', 'fraldas', 'cerveja']
]
```

```python
#Aplicando apriori e salvando os resultados em uma variavel chamada regras
regras = list(apriori(transacoes, min_support=0.1, min_confidence=0.8, min_lift=1.5))
```

```python
# codigo que navega no resultado salvo na variavel regras, e deixa as informacoes organizadas para ter uma melhor visualizacao
for registro in regras:
    # Conjunto de Itens
    itemset = registro.items

    # Suporte do Itemset
    support = registro.support

    print(f"Conjunto de Itens: {itemset}")
    print(f"Suporte: {support}")

    # Estatísticas Ordenadas
    for stat in registro.ordered_statistics:
        print("\nEstatísticas Ordenadas:")
        print(f"  Itens Base: {stat.items_base}")
        print(f"  Itens Adicionados: {stat.items_add}")
        print(f"  Confiança: {stat.confidence}")
        print(f"  Lift: {stat.lift}")

    print("\n" + "-"*30)
```

```python
# aqui, vamos pegar o csv gerado no SQL, e usando o pandas para criar um dataframe
datasetVendas = pd.read_csv('/content/drive/MyDrive/TCC/Dados_Vendas.csv', delimiter=';')
```

```python
# visualizando os dados do dataframe, para saber se esta tudo correto
datasetVendas
```

```python
# Transformar os dados para o formato esperado pelo Apriori
Conjunto_Vendas = datasetVendas.groupby('ID_Venda')['Nome_Produto'].apply(list).values.tolist()
```

```python
# visualizando os dados do dataframe, para saber se esta tudo correto
Conjunto_Vendas
```

```python
#Aplicando apriori e salvando os resultados em uma variavel chamada regras
regras = list(apriori(Conjunto_Vendas, min_support=0.003, min_confidence=0.8, min_lift=1.5))
```

```python
# codigo que navega no resultado salvo na variavel regras, e deixa as informacoes organizadas para ter uma melhor visualizacao
for registro in regras:
    # Conjunto de Itens
    itemset = registro.items

    # Suporte do Itemset
    support = registro.support

    print(f"Conjunto de Itens: {itemset}")
    print(f"Suporte: {support}")

    # Estatísticas Ordenadas
    for stat in registro.ordered_statistics:
        print("\nEstatísticas Ordenadas:")
        print(f"  Itens Base: {stat.items_base}")
        print(f"  Itens Adicionados: {stat.items_add}")
        print(f"  Confiança: {stat.confidence}")
        print(f"  Lift: {stat.lift}")

    print("\n" + "-"*30)
```

```python
# Agrupando os produtos por venda, mês e criando listas de produtos
Conjunto_Mes = datasetVendas.groupby(['ID_Venda', 'Mes'])['Nome_Produto'].apply(list)

# Adicionando o mês à lista de produtos
Conjunto_Mes = Conjunto_Mes.reset_index()
Conjunto_Mes['Vendas'] = Conjunto_Mes.apply(lambda row: [row['Mes']] + row['Nome_Produto'], axis=1)

# Criando a lista final de transações
Conjunto_Com_Mes = Conjunto_Mes['Vendas'].apply(set).tolist()
```

```python
# visualizando os dados do dataframe, para saber se esta tudo correto
Conjunto_Com_Mes
```

```python
#Aplicando apriori e salvando os resultados em uma variavel chamada regras_mes
regras_mes = list(apriori(Conjunto_Com_Mes, min_support=0.03, min_confidence=0.8, min_lift=1.5))
```

```python
# codigo que navega no resultado salvo na variavel regras, e deixa as informacoes organizadas para ter uma melhor visualizacao
for registro in regras_mes:
    # Conjunto de Itens
    itemset = registro.items

    # Suporte do Itemset
    support = registro.support

    print(f"Conjunto de Itens: {itemset}")
    print(f"Suporte: {support}")

    # Estatísticas Ordenadas
    for stat in registro.ordered_statistics:
        print("\nEstatísticas Ordenadas:")
        print(f"  Itens Base: {stat.items_base}")
        print(f"  Itens Adicionados: {stat.items_add}")
        print(f"  Confiança: {stat.confidence}")
        print(f"  Lift: {stat.lift}")

    print("\n" + "-"*30)
```
